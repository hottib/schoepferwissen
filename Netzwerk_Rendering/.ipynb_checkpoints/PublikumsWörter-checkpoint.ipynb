# Publikums Wörter

```python
import pandas as pd
from datetime import datetime
print("Imports done!")


publikum_str = """
Wahrheit
Körper 
Seele
Herz
Frei
Innen
Kraft
Lügen
Organe
Tun
Emotionen
Secret
Kontrolle
Kampf
gefühlt
Stimme
Warheit
Mensch
Seele
Leben
keinen
Nichts
Menschen
Herzschlag
schlag
Programmierung
Stille
Körpe
können
Mesias
"""
pub_str_lower = publikum_str.lower()
pub_list = pub_str_lower.split()

df = pd.read_csv("Node300.csv")
df['Color'] = "not_selected"

filt = df["Label"].isin(pub_list)
pub_df = df[filt]

counter = 0
for Id in pub_df["Id"]:
    df["Color"][Id] = "selected"
    counter +=1
    print("Hit")
print("Treffer: %s" %counter)
    #Save
now = datetime.now()
time = now.strftime("%d_%H-%M")
time_str = str(time)

filenamePub = "Node_Publikum_Selection__"+ time +".csv"
df.to_csv(filenamePub, index=False)
print("CSV saved as: %s" %filenamePub)
```
